```python
# Return Object to Float but if there is any k present it convert it into 1000
```

```python
def replace_alphabets_from_values(v: str) -> float :
    """ Return the mathematical values of some values like (1.3k -> 1300) """
    try:
        value = float(v)
    except ValueError:
        value = float(v.replace('k','')) * 1000
    return value
```

```python
df['followings'] = df['followings'].apply(replace_alphabets_from_values)
```

```python
df['joining_year'].dtype
```

```python
df.stars = df.stars.apply(replace_alphabets_from_values).astype('int32')
```

```python
df.repositories = df.repositories.apply(replace_alphabets_from_values).astype('int')
```

```python
df.followers = df.followers.apply(replace_alphabets_from_values).astype('int32')
```
